```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 100)

# this assumes your CSV is in the SAME folder as this notebook
RAW_DATA_PATH = "US_Accidents_March23.csv"   # <-- change this to your actual file name
```

```python
preview = pd.read_csv(RAW_DATA_PATH, nrows=5)
print(list(preview.columns))
print(preview.shape)
```

```python
import os

file_size_gb = os.path.getsize(RAW_DATA_PATH) / (1024**3)
print(f"File size: {file_size_gb:.2f} GB")

with open(RAW_DATA_PATH, "r", encoding="utf-8", errors="ignore") as f:
    total_rows = sum(1 for _ in f) - 1  # minus header row
print(f"Total rows: {total_rows:,}")
```

```python
TOTAL_ROWS = 7_728_394          # from what you just found
TARGET_SAMPLE_SIZE = 500_000
SAMPLE_FRACTION = TARGET_SAMPLE_SIZE / TOTAL_ROWS
CHUNK_SIZE = 200_000

sampled_chunks = []
chunk_num = 0

for chunk in pd.read_csv(RAW_DATA_PATH, chunksize=CHUNK_SIZE, low_memory=False):
    chunk_num += 1
    piece = chunk.sample(frac=SAMPLE_FRACTION, random_state=42)
    sampled_chunks.append(piece)
    print(f"Processed chunk {chunk_num}, collected so far: {sum(len(c) for c in sampled_chunks):,} rows")

df_sample = pd.concat(sampled_chunks, ignore_index=True)
print("\nFinal sample shape:", df_sample.shape)
```

```python
df_sample.to_parquet("accidents_sample.parquet", index=False)
print("Saved! File size:", os.path.getsize("accidents_sample.parquet") / (1024**2), "MB")
```

```python
print("Shape:", df_sample.shape)
print("\n--- Data types ---")
print(df_sample.dtypes)
print("\n--- Missing values (top 20, %) ---")
missing_pct = (df_sample.isnull().sum() / len(df_sample) * 100).sort_values(ascending=False)
print(missing_pct.head(20))
```

```python
# check for zero-variance boolean columns first
bool_cols = df_sample.select_dtypes(include='bool').columns
for col in bool_cols:
    print(col, "->", df_sample[col].value_counts(normalize=True).to_dict())
```

```python
df_clean = df_sample.copy()

# 1. Drop: structurally missing, redundant/derived, pure metadata, zero/near-zero variance, identifiers
cols_to_drop = [
    'End_Lat', 'End_Lng',                                              # structurally missing
    'Wind_Chill(F)',                                                   # derived from Temp + Wind_Speed
    'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight',    # redundant w/ Sunrise_Sunset
    'Weather_Timestamp', 'Airport_Code',                               # metadata about the weather station
    'Turning_Loop',                                                    # 100% one value
    'Roundabout',                                                      # 99.996% one value
    'ID',                                                              # unique identifier
]
df_clean = df_clean.drop(columns=cols_to_drop)

# 2. Precipitation missing = no precipitation, not "unknown"
df_clean['Precipitation(in)'] = df_clean['Precipitation(in)'].fillna(0)

# 3. Small sensor gaps in numeric weather columns -> median impute
numeric_impute_cols = ['Temperature(F)', 'Humidity(%)', 'Pressure(in)',
                        'Visibility(mi)', 'Wind_Speed(mph)']
for col in numeric_impute_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())

# 4. Categorical weather gaps -> explicit "Unknown" category
df_clean['Wind_Direction'] = df_clean['Wind_Direction'].fillna('Unknown')
df_clean['Weather_Condition'] = df_clean['Weather_Condition'].fillna('Unknown')

# 5. Negligible missing (<0.3%) on remaining columns -> just drop those rows
df_clean = df_clean.dropna(subset=['Street', 'Timezone', 'Zipcode', 'Sunrise_Sunset'])

print("Shape before cleaning:", df_sample.shape)
print("Shape after cleaning:", df_clean.shape)
print("\nAny missing values left?")
remaining = df_clean.isnull().sum()
print(remaining[remaining > 0] if remaining.sum() > 0 else "None — fully clean.")
```

```python
# 1. Binary target — what we decided early on, locked in now
df_clean['Risk_Target'] = (df_clean['Severity'] >= 3).astype(int)

# 2. Time features — Hour and Month kept as-is (we proved both carry real signal)
# Weekday as a number (Monday=0) is fine for tree models, no need to keep text version
df_clean['Weekday_Num'] = df_clean['Start_Time'].dt.dayofweek  # 0=Monday, 6=Sunday
df_clean['Is_Weekend'] = (df_clean['Weekday_Num'] >= 5).astype(int)

# 3. Rush hour flag — built carefully based on what we actually found, not a generic assumption
# We saw BOTH morning (7-9) and evening (16-19) rush windows elevated, plus a separate 
# late-night/early-morning spike (4-6 AM) that is NOT "rush hour" but is also high-risk.
# We'll keep Hour itself as the main signal (already proven informative) and add 
# Rush_Hour as a supplementary flag matching what the data actually showed, not textbook rush hour times.
df_clean['Is_Rush_Hour'] = df_clean['Hour'].isin([7,8,9,16,17,18,19]).astype(int)
df_clean['Is_Late_Night'] = df_clean['Hour'].isin([0,1,2,3,4,5]).astype(int)

print(df_clean[['Severity','Risk_Target','Hour','Weekday_Num','Is_Weekend','Is_Rush_Hour','Is_Late_Night','Month']].head(10))
print("\nRisk_Target distribution:")
print(df_clean['Risk_Target'].value_counts(normalize=True).round(4) * 100)
```

```python
# CRITICAL STEP: leakage check
# For every column, ask: "would I know this BEFORE the accident's severity is determined,
# or only AFTER, as a consequence of it?" Anything in the second group gets dropped.

leakage_cols = [
    'End_Time',        # only known after the incident is fully cleared
    'Distance(mi)',    # length of road affected -- a CONSEQUENCE of severity, not a cause
    'Description',     # free text written about the incident, often describing the outcome
]

redundant_or_noisy_cols = [
    'Severity',         # this IS our target -- raw version must not leak into the features
    'Start_Time',       # already extracted Hour, Month, Weekday_Num from this
    'Weekday',          # text version -- we kept Weekday_Num instead
    'Street', 'City', 'County', 'Zipcode',  # too granular, near-unique per row, won't generalize
    'Country',          # check below if this even varies
    'Timezone',         # redundant with State
]

print("Unique values in Country:", df_clean['Country'].unique())

df_features = df_clean.drop(columns=leakage_cols + redundant_or_noisy_cols)
print("\nShape after dropping leakage/redundant columns:", df_features.shape)
print("\nRemaining columns:")
print(list(df_features.columns))
```

```python
# Group rare weather conditions into 'Other' before one-hot encoding
weather_counts = df_features['Weather_Condition'].value_counts()
common_weather = weather_counts[weather_counts >= 1000].index
df_features['Weather_Condition'] = df_features['Weather_Condition'].apply(
    lambda x: x if x in common_weather else 'Other'
)
print("Weather categories after grouping:", df_features['Weather_Condition'].nunique())

# Same treatment for Wind_Direction (it has near-duplicate values like 'CALM' vs 'Calm')
df_features['Wind_Direction'] = df_features['Wind_Direction'].str.upper()
wind_counts = df_features['Wind_Direction'].value_counts()
common_wind = wind_counts[wind_counts >= 1000].index
df_features['Wind_Direction'] = df_features['Wind_Direction'].apply(
    lambda x: x if x in common_wind else 'Other'
)
print("Wind direction categories after grouping:", df_features['Wind_Direction'].nunique())

# One-hot encode the grouped weather/wind columns
df_encoded = pd.get_dummies(df_features, columns=['Weather_Condition', 'Wind_Direction'], prefix=['Weather', 'Wind'])

# Label-encode State and Source (safe for tree-based models)
from sklearn.preprocessing import LabelEncoder
le_state = LabelEncoder()
le_source = LabelEncoder()
df_encoded['State_Encoded'] = le_state.fit_transform(df_encoded['State'])
df_encoded['Source_Encoded'] = le_source.fit_transform(df_encoded['Source'])
df_encoded = df_encoded.drop(columns=['State', 'Source'])

# Binary encode Sunrise_Sunset
df_encoded['Is_Day'] = (df_encoded['Sunrise_Sunset'] == 'Day').astype(int)
df_encoded = df_encoded.drop(columns=['Sunrise_Sunset'])

print("\nFinal shape:", df_encoded.shape)
print("\nAll dtypes:")
print(df_encoded.dtypes.value_counts())
```

```python
df_encoded.to_parquet("accidents_features.parquet", index=False)
print("Saved! Shape:", df_encoded.shape)
```

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop(columns=['Risk_Target'])
y = df_encoded['Risk_Target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("\nTrain class balance:")
print(y_train.value_counts(normalize=True).round(4) * 100)
print("\nTest class balance:")
print(y_test.value_counts(normalize=True).round(4) * 100)
```

```python
from sklearn.ensemble import RandomForestClassifier
import time

rf_model = RandomForestClassifier(
    n_estimators=200,        # number of trees -- more trees = more stable, diminishing returns past ~200-300
    max_depth=20,             # caps tree depth to avoid severe overfitting on 73 features
    min_samples_leaf=20,      # each leaf needs at least 20 samples -- smooths out noisy splits
    class_weight='balanced',  # the imbalance fix we just discussed
    n_jobs=-1,                 # use all CPU cores -- this matters a lot at 400k rows
    random_state=42,
    verbose=1
)

print("Training Random Forest...")
start = time.time()
rf_model.fit(X_train, y_train)
elapsed = time.time() - start
print(f"\nTraining completed in {elapsed:.1f} seconds")
```

```python
from sklearn.metrics import classification_report

y_pred_rf = rf_model.predict(X_test)
print(classification_report(y_test, y_pred_rf, target_names=['Low-Risk','High-Risk']))
```

```python
import joblib
joblib.dump(rf_model, 'rf_model.joblib')
print("Random Forest saved to disk.")
```

```python
plt.figure(figsize=(8,7))
for name, probs in models_probs.items():
    precision, recall, _ = precision_recall_curve(y_test, probs)
    ap = average_precision_score(y_test, probs)
    plt.plot(recall, precision, label=f'{name} (AP={ap:.3f})')

baseline = y_test.mean()  # what a "no-skill" model achieves -- just the positive class rate
plt.axhline(baseline, color='k', linestyle='--', label=f'No-skill baseline ({baseline:.3f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves - Model Comparison')
plt.legend()
plt.savefig('pr_curves.png', dpi=100, bbox_inches='tight')
plt.show()

print("Average Precision (PR-AUC) Scores:")
for name, probs in models_probs.items():
    ap = average_precision_score(y_test, probs)
    print(f"{name}: {ap:.4f}")
```

```python
precision_xgb, recall_xgb, thresholds_xgb = precision_recall_curve(y_test, y_prob_xgb)

# F1 balances precision and recall equally -- compute it at every threshold and find the peak
f1_scores = 2 * (precision_xgb * recall_xgb) / (precision_xgb + recall_xgb + 1e-10)
best_idx = np.argmax(f1_scores)
best_threshold = thresholds_xgb[best_idx]

print(f"Best threshold (max F1): {best_threshold:.3f}")
print(f"At this threshold -> Precision: {precision_xgb[best_idx]:.3f}, Recall: {recall_xgb[best_idx]:.3f}, F1: {f1_scores[best_idx]:.3f}")

# compare directly against the default 0.5 cutoff
from sklearn.metrics import f1_score
y_pred_default = (y_prob_xgb >= 0.5).astype(int)
print(f"\nAt default 0.5 threshold -> F1: {f1_score(y_test, y_pred_default):.3f}")
```

```python
import pandas as pd
import numpy as np
import json
import os

# 1. Lightweight sample for the heatmap (20k rows, not 500k -- keeps the map snappy in-browser)
df_raw_sample = pd.read_parquet("accidents_sample.parquet")
df_raw_sample['Risk_Target'] = (df_raw_sample['Severity'] >= 3).astype(int)

dashboard_sample = df_raw_sample[[
    'Start_Lat', 'Start_Lng', 'State', 'Weather_Condition',
    'Severity', 'Risk_Target', 'Junction', 'Traffic_Signal', 'Crossing', 'Stop'
]].sample(20000, random_state=42)
dashboard_sample.to_parquet("dashboard_sample.parquet", index=False)

# 2. Precomputed aggregate tables -- same logic as our EDA, just saved for instant reuse
state_risk_table = df_raw_sample.groupby('State').agg(
    Total_Accidents=('Severity', 'count'), High_Severity_Rate=('Risk_Target', 'mean')
).reset_index()
state_risk_table['High_Severity_Rate'] = (state_risk_table['High_Severity_Rate'] * 100).round(2)
state_risk_table = state_risk_table[state_risk_table['Total_Accidents'] >= 500]
state_risk_table.to_json('state_risk_table.json', orient='records')

weather_counts = df_raw_sample['Weather_Condition'].value_counts()
reliable_weather = weather_counts[weather_counts >= 500].index
weather_risk_table = df_raw_sample[df_raw_sample['Weather_Condition'].isin(reliable_weather)].groupby('Weather_Condition').agg(
    Total_Accidents=('Severity', 'count'), High_Severity_Rate=('Risk_Target', 'mean')
).reset_index()
weather_risk_table['High_Severity_Rate'] = (weather_risk_table['High_Severity_Rate'] * 100).round(2)
weather_risk_table = weather_risk_table.sort_values('High_Severity_Rate', ascending=False)
weather_risk_table.to_json('weather_risk_table.json', orient='records')

road_cols = ['Junction', 'Traffic_Signal', 'Crossing', 'Stop', 'Station', 'Amenity']
road_risk_table = pd.DataFrame({
    'Feature': road_cols,
    'Severity_When_Present': [(df_raw_sample.loc[df_raw_sample[c]==True, 'Risk_Target'].mean()*100) for c in road_cols],
    'Severity_When_Absent': [(df_raw_sample.loc[df_raw_sample[c]==False, 'Risk_Target'].mean()*100) for c in road_cols],
})
road_risk_table.to_json('road_risk_table.json', orient='records')

# 3. Global SHAP importance -- reusing the shap_values we already calculated, NOT recomputing
mean_abs_shap = np.abs(shap_values).mean(axis=0)
feature_importance = pd.DataFrame({
    'Feature': X_test_sample.columns, 'Mean_Abs_SHAP': mean_abs_shap
}).sort_values('Mean_Abs_SHAP', ascending=False).head(20)
feature_importance.to_json('feature_importance.json', orient='records')

print("Dashboard data files ready:")
for f in ['dashboard_sample.parquet', 'state_risk_table.json', 'weather_risk_table.json',
          'road_risk_table.json', 'feature_importance.json', 'deployment_metadata.json',
          'model_config.json', 'xgb_model.joblib']:
    print(f"  {'✓' if os.path.exists(f) else '✗ MISSING'}  {f}")
```
